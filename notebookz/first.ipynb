# #First steps with Polars

```python
# 3rd party imports
import polars as pl
import datetime as dt
```

```python

df = pl.DataFrame(
    {
        "name": ["Alice Archer", "Ben Brown", "Bucky Richard Donald", "Chloe Cooper", "Daniel Donovan", "Eddie Ernest" ],
        "birthdate": [
            dt.date(1997, 1, 10),
            dt.date(1985, 2, 15),
            dt.date(1949, 5, 23),
            dt.date(1983, 3, 22),
            dt.date(1981, 4, 30),
            dt.date(1977, 11, 23),
        ],
        "weight": [57.9, 72.5, 66.42, 53.6, 83.1, 66.6],  # (kg)
        "height": [1.56, 1.77, 1.65, 1.68, 1.75, 1.84],  # (m)
    }
)

df

```

```python
result = df.select(
    pl.col(["name", "weight"]),
    pl.col("birthdate").dt.year().alias("birth_year"),
    (pl.col("weight") / (pl.col("height") ** 2)).alias("bmi"),
)
result
```
