```python
import matplotlib.pyplot as plt
```

```python
data = df.group_by("Alignment").len() # #value_counts()
color = ["red","blue", "yellow", "green"]

plt.subplots(figsize =(10,10))
plt.pie(data["len"],labels=data["Alignment"],autopct ="%1.1f%%", colors=color)
plt.title("Alignments")
plt.legend()
data
```

```python
data = df.group_by("Species").len() #
data.filter(pl.col("Species")!="-").filter(pl.col("Species")!="Human")
```

```python
exlude = ["Human / Cosmic","Human / Radiation", "Human", "Human / Radiation", "Human / Altered" , "Human / Radiation", "Human-Spartoi", "Human-Kree"]
df = df.filter(pl.col("Species")!="-").filter(~pl.col("Species").is_in(exlude))
data = df.group_by("Species").len() # #value_counts()


plt.subplots(figsize =(10,10))
plt.pie(data["len"],labels=data["Species"],autopct ="%1.1f%%")
plt.title("Species != Human")
#plt.legend()
print(df['Species'].unique())
```

```python
df = pl.read_csv("../dta/marvel_sanitized_w_h.csv")
include = ["Human / Cosmic","Human / Radiation", "Human", "Human / Radiation", "Human / Altered" , "Human / Radiation", "Human-Spartoi", "Human-Kree"]
df = df.filter(pl.col("Species")!="-").filter(pl.col("Species").is_in(include))
data = df.group_by("Species").len() # #value_counts()


plt.subplots(figsize =(10,10))
plt.pie(data["len"],labels=data["Species"],autopct ="%1.1f%%")
plt.title("Species == Human*")
#plt.legend()
print(df['Species'].unique())
```
